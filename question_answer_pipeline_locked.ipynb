```python
# importing piprline
from transformers import pipeline
```

```python
# making instance of pipeline.
question_answer = pipeline("question-answering")
```

```python
# making simple question and context with the instance of the pipeline to get answer.
question_answer ( question = "where do you work?",
                 context = "my name is Atif i work at hugging face at pakistan.")
```

```python
# real fun.
# creating custom dataset with two columns such as qoestion and context.
import pandas as pd

data = {
    "question": [
        "Where do you work?",
        "What is your favorite hobby?",
        "Can you recommend a good book?",
        "Tell me about your educational background.",
        "What's your favorite cuisine?",
        "Do you enjoy traveling?",
        "How do you spend your weekends?",
        "What languages do you speak?",
        "Tell me about your family.",
        "What's your dream job?",
        "Have you ever met a celebrity?",
        "What's the best movie you've seen recently?",
        "What's your favorite music genre?",
        "Do you follow any sports?",
        "Tell me about your pets, if any.",
        "What's your most memorable life experience?",
        "What are your future goals?",
        "How do you handle stress?",
        "Tell me about a skill you're proud of.",
        "What's your favorite season of the year?"
    ],
    "context": [
        "My name is Atif. I work at Hugging Face in Pakistan.",
        "I enjoy hiking and painting in my free time.",
        "Sure! I recently read 'The Alchemist' by Paulo Coelho.",
        "I have a degree in Computer Science from XYZ University.",
        "I love trying different types of sushi.",
        "Absolutely! I've traveled to 10 different countries.",
        "I usually relax at home and catch up on reading.",
        "I'm fluent in English, French, and Spanish.",
        "I come from a close-knit family of five.",
        "My dream job is to become an astronaut.",
        "Yes, I met a famous actor at a charity event.",
        "I was blown away by 'Inception'.",
        "I'm a big fan of rock music.",
        "I'm a die-hard football (soccer) fan.",
        "I have a playful Labrador named Max.",
        "Skydiving was definitely a thrilling experience.",
        "I plan to write a novel and travel the world.",
        "I practice meditation and deep breathing.",
        "I'm proud of my photography skills.",
        "I love the coziness of winter."
    ]
}

df = pd.DataFrame(data)

df
```

```python
# to read each qestion and context of df i used for loop.
# basically with help of index it reaches at every row and read that row data mean print that . it run till end of df length.
for index, row in df.iterrows():
    question = row["question"]
    context = row["context"]
    print(f"Question: {question}")
    print(f"Context: {context}")
    print("-" * 30)
```

```python
# here i empty list is created to store the answers which are given by pipeline of of question_answer.
# firstly those ansers are saved in answers variable than append to the actual list.
answer_list = []
for index, row in df.iterrows():
    question = row["question"]
    context = row["context"]
    answers = question_answer(question, context)
    answer_list.append(answers)
```

```python
# the answers are in from of dictionary having much information and we need only the answer.
answer_list
```

```python
# to get the important information from the answer_list which is answer.
# so again a empty list <filter_answer_list> is created than run the for loop in answer_list than append the ans in filter_answer_list
filter_answer_list=[]
for ans in answer_list:
  filter_answer_list.append(ans["answer"])
```

```python
# this is what is required answer of question and context of df .
filter_answer_list
```

```python
len(filter_answer_list)
```

```python
# making a column in df called answers to show the answer of every question and context.
df['answers'] = filter_answer_list
```
