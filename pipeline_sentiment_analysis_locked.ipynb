```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier("Today was an incredible day! I aced my presentation and received praise from my colleagues. Feeling on top of the world right now! 😄🎉")
```

```python
import pandas as pd
data = {
    'Sentiment': ["Today was an incredible day! I aced my presentation and received praise from my colleagues. Feeling on top of the world right now! 😄🎉",
        "Absolutely loved the new restaurant in town. The food was exceptional, and the ambiance was perfect for a cozy dinner. Highly recommended!",
        "Just adopted the most adorable puppy. My heart is melting every time it cuddles up to me. Can't wait for all the new adventures together! 🐶❤️",
        "Finished reading a fantastic book that left me completely engrossed. The author's storytelling skills are beyond impressive. Can't wait to explore more of their work!",
        "Spent the day at the beach with friends, basking in the sun and enjoying the waves. Nothing beats the feeling of sand between your toes and laughter in the air. ☀️🏖️",
        "Today was just one of those days. Everything seemed to go wrong, from missing my bus to spilling coffee all over my work. Can't wait for this day to end.",
        "Disappointed by the service at the new cafe. The staff was rude, and the food took forever to arrive. Won't be going back there anytime soon.",
        "Lost my wallet today, along with all my important cards and cash. Feeling frustrated and stressed about the hassle of replacing everything.",
        "Tried watching a highly recommended movie, but it turned out to be a complete bore. Two hours of my life I'll never get back.",
        "Had to deal with a flat tire on my way to an important meeting. Just when you think things can't get worse, they do."],
    'Sentiment Type': [
        'Positive', 'Positive', 'Positive', 'Positive', 'Positive',
        'Negative', 'Negative', 'Negative', 'Negative', 'Negative'
    ]
}
df = pd.DataFrame(data)
df
```

```python
for i in df.Sentiment:
  results = classifier(i)
  print(results)
```
