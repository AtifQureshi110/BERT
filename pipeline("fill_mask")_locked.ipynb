- masker = pipeline("fill-mask") :

- The pipeline("fill-mask") creates a pipeline for masked language modeling tasks.
- This pipeline finds missing words based on the model and using mathematical and computational calculations.
- It requires a text input with a placeholder like <mask> to indicate the missing word's position.
- The top_k parameter (not explicitly mentioned in your description but inferred from context) determines how many predictions to return from the pipeline.

#### pipeline


```python
from transformers import pipeline
```

```python
masker = pipeline("fill-mask")
```

#### masker

```python
input = "this course will teach you all about <mask> model."
top_predications = masker(input, top_k=2)
```

```python
top_predications
```

```python
for top_predication in top_predications:
  print(top_predication['token_str'], top_predication['score'])
  # Compare scores to find the highest one
```

```python
highest_score = -1
selected_token = None
for top_predication in top_predications:
  token = top_predication['token_str']
  score = top_predication['score']
  if score > highest_score:
    highest_score = score
    selected_token = token
print("Selected Token:", selected_token)
print("Highest Score:", highest_score)
```

```python
replaced_sentence = input.replace("<mask>", selected_token)
# Print the replaced sentence
print("Original Sentence: ",input)
print("Replaced Sentence: ", replaced_sentence)
```

```python
import pandas as pd

# Sample sentences with masked words
sentences = [
    "This course will teach you all about <mask> models.",
    "I can't believe how <mask> this is!",
    "The key to success is <mask> and hard work.",
    "The world is full of <mask> opportunities.",
    "Learning to <mask> is essential in today's world.",
    "The recipe calls for flour, sugar, and <mask>.",
    "His new novel is a <mask> of suspense and mystery.",
    "She has a <mask> for playing musical instruments.",
    "The conference will focus on <mask> advancements.",
    "Innovation is driven by <mask> and curiosity.",
    "The artist's <mask> are on display at the gallery.",
    "Nature provides us with <mask> resources.",
    "The team's success is a result of <mask> effort.",
    "The technology has the potential to <mask> industries.",
    "The path to success is paved with <mask> work.",
    "The scientific method involves observation and <mask>.",
    "Her speech was met with <mask> applause.",
    "The company's values are <mask> in its mission statement.",
    "He has a talent for solving <mask> problems.",
    "The universe is full of <mask> mysteries."
]

# Create a DataFrame
df = pd.DataFrame(sentences, columns=['Masked Sentence'])

# Display the DataFrame
df
```

#### Predication with top_k=2

```python
# here it show all information of each
for i in df["Masked Sentence"]:
  top_pred = masker(i, top_k=2)
  print(top_pred)
```

```python
# Initialize lists to store prediction data
predicted_words = []
predicted_scores = []
predicted_sequences = []
for i in df["Masked Sentence"]:
  top_pred = masker(i, top_k=2)
  for prediction in top_pred:
    print(f"{prediction['token_str']},\t {prediction['score']},\t {prediction['sequence']}")
    predicted_words.append(prediction['token_str'])
    predicted_scores.append(prediction['score'])
    predicted_sequences.append(prediction['sequence'])
```

```python
# Create a DataFrame to store the prediction data
prediction_df = pd.DataFrame({
    'Predicted Word': predicted_words,
    'Score': predicted_scores,
    'Sequence': predicted_sequences
})

# Print the prediction DataFrame
prediction_df.head(50)
```

#### Prediction with top_k, selecting the most valuable word among the predicted options.

```python
# Initialize lists to store prediction data
predicted_words_0 = []
predicted_scores_0 = []
predicted_sequences_0 = []

for i in df["Masked Sentence"]:
    top_pred = masker(i, top_k=2)

    highest_score = -1
    selected_prediction = None

    for prediction in top_pred:
        score = prediction['score']

        if score > highest_score:
            highest_score = score
            selected_prediction = prediction
            print(f"{prediction['token_str']},\t {prediction['score']},\t {prediction['sequence']}")

    if selected_prediction:
        predicted_words_0.append(selected_prediction['token_str'])
        predicted_scores_0.append(selected_prediction['score'])
        predicted_sequences_0.append(selected_prediction['sequence'])
```

```python
# Create a DataFrame to store the prediction data
prediction_df = pd.DataFrame({'Predicted Word': predicted_words_0, 'Score': predicted_scores_0, 'Sequence': predicted_sequences_0})
# Print the prediction DataFrame
prediction_df.head(50)
```

```python
masked_df = pd.concat([df, prediction_df], axis=1)
```
